# Gridsearch

The purpose of this section is to find the best local model for each aquiferes (ie 6 models). Those models can have different classes as well as different hyperparameters.

Let us start with TARSAC (the toughest one)

I'll try:
- FourTheta
- Linear Regression with and without covariates
- LGBM with and without covariates
- LSTM with covariates


```python
from darts.utils import
```

```python
# gridsearch config
start=pd.Timestamp('2015-06-01')
forecast_horizon=182
metric=mape
stride=365
location = 'saint_porquier'
```

```python
# Exponential Smoothing
from darts.models import ExponentialSmoothing
model_exp = ExponentialSmoothing()
parameters = {'seasonal_periods': [2, 4]}
model_exp.gridsearch(parameters = parameters, series = targets[location], 
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
# FourTheta
model_fourtheta = FourTheta()

parameters = {
    'seasonality_period': [30, 365],
}
warnings.filterwarnings("ignore")
model_fourtheta.gridsearch(parameters = parameters, series = targets[location], 
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
#Linear regression without covariates
model_linear_nocov = LinearRegressionModel(lags=1, output_chunk_length=180)
parameters = {
    'lags': [1, 10, 20, 30, 45, 90, 180, 360],
    'output_chunk_length': [1, 10, 20, 45, 90, 180]
}
model_linear_nocov.gridsearch(parameters = parameters, series = targets[location], 
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
#Linear regression with 3 covariates
features = ['pluie', 'etp', 'pompage']

targets, covs = build_target_and_covs(compact_df_without_missing, features=features)
model_linear_cov = LinearRegressionModel(lags=1, output_chunk_length=180, lags_past_covariates=1)
parameters = {
    'lags': [10, 20, 50, 180, 365],
    'output_chunk_length': [180],
    'lags_past_covariates': [10, 20, 50, 180, 365],
}
model_linear_cov.gridsearch(parameters = parameters, series = targets[location], past_covariates=covs[location],
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
#Linear regression with 7 covariates
features = ['pluie', 'etp', 'pompage', 'pluie_annuelle', 'pluie_mensuelle', 'etp_annuelle', 'etp_mensuelle']

targets, covs = build_target_and_covs(compact_df_without_missing, features=features)


model_linear_cov = LinearRegressionModel(lags=1, output_chunk_length=180, lags_past_covariates=1)
parameters = {
    'lags': [10, 20, 50, 180, 365],
    'output_chunk_length': [180],
    'lags_past_covariates': [10, 20, 50, 180, 365],
}
model_linear_cov.gridsearch(parameters = parameters, series = targets[location], past_covariates=covs[location],
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
features = ['pluie', 'etp', 'pompage', 'pluie_annuelle', 'pluie_mensuelle', 'etp_annuelle', 'etp_mensuelle', 
           'pluie_lag_Y-1', 'pluie_lag_Y-2', 'etp_lag_Y-1', 'etp_lag_Y-2']

targets, covs = build_target_and_covs(compact_df_without_missing, features=features)


#Linear regression with 11 covariates

model_linear_cov = LinearRegressionModel(lags=1, output_chunk_length=180, lags_past_covariates=1)
parameters = {
    'lags': [10, 20, 50, 180, 365],
    'output_chunk_length': [180],
    'lags_past_covariates': [10, 20, 50, 180, 365],
}
model_linear_cov.gridsearch(parameters = parameters, series = targets[location], past_covariates=covs[location],
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
from darts.models import ARIMA

# ARIMA
model_arima = ARIMA()
parameters = {
    'p': [2],
    'd': [0],
    'q': [2]
}
model_arima.gridsearch(parameters = parameters, series = targets[location],
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
# way to long
from darts.models import LightGBMModel
lgb_model = LightGBMModel(lags=60, output_chunk_length=10)
parameters = {
    'lags': [1, 10, 20, 30],
    'output_chunk_length': [1, 10, 20, 45, 90, 180]
}
lgb_model.gridsearch(parameters = parameters, series = targets[location],
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
lgb_model = LightGBMModel(lags=30, output_chunk_length=180, lags_past_covariates=1)
parameters = {
    'lags_past_covariates': [1, 10, 20, 30, 60],
}
lgb_model.gridsearch(parameters = parameters, series = targets[location], past_covariates=covs[location],
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```

```python
# LSTM => sucks
from darts.models import RNNModel
lstm_model = RNNModel(model='LSTM', input_chunk_length=1)
parameters = {
    'input_chunk_length': [10, 20, 50],
    'training_length': [25, 50, 100],
    'n_rnn_layers': [1, 2, 3],
    'hidden_dim': [32, 64, 128, 256]
}
parameters = {
    'input_chunk_length': [20]
}
lstm_model.gridsearch(parameters = parameters, series = targets[location], past_covariates=covs[location],
                          start=start, forecast_horizon=forecast_horizon, metric=mape, stride=stride)
```
