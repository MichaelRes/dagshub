```python
import darts
from darts import TimeSeries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("groundwater_prediction_dataset/dataset.csv", sep=";")
```

```python
models = data["model_name"].unique().tolist()
models
```

```python
def compact_dataset():
    dates = data["Date"].unique()
    models = ['bioule', 'les_barthes', 'saint_felix', 'saint_porquier', 'tarsac', 'verniolles']
    features = ['pz', 'pluie', 'etp', 'pompage']
    compact_data = np.zeros((len(dates) * len(models), 6),  dtype=object)
    for i in range(len(dates)):
        if i%100 == 0:
            print(f'{i}/{len(dates)} dates done')
        data_row = data[data["Date"] == dates[i]]
        
        for j in range(len(models)):
            data_row_local = data_row[data_row['model_name'] == models[j]]
            j_ = i * len(models) + j
            compact_data[j_][0] = dates[i]
            compact_data[j_][1] = models[j]
            for k in range(len(features)):
                try:
                    compact_data[j_][2 + k] = data_row_local.loc[data_row_local['variable'] == features[k], 'values'].values[0]
                except:
                    compact_data[j_][2 + k] = np.nan
    
    compact_df = pd.DataFrame(compact_data, columns=['Date', 'model_name', 'pz', 'pluie', 'etp', 'pompage'])
    return compact_df
    
```

```python
# compact_df = compact_dataset()

compact_df = pd.read_csv("compact_df.csv", index_col = 0)
compact_df_without_missing = compact_df[~compact_df['pz'].isnull()]
compact_df_without_missing['Date'] = pd.to_datetime(compact_df_without_missing['Date'], format='%Y-%m-%d')
```

```python
from darts.metrics import mape, smape, mase, mae, rmse

def print_metrics(validation, pred, train):
    MAPE = mape(validation, pred)
    sMAPE = smape(validation, pred)
    MASE = mase(validation, pred, train)
    MAE = mae(validation, pred)
    RMSE = rmse(validation, pred)
    print(f'MAPE: {MAPE}')
    print(f'sMAPE: {sMAPE}')
    print(f'MASE: {MASE}')
    print(f'MAE: {MAE}')
    print(f'RMSE: {RMSE}')
    
    return [MAPE, sMAPE, MASE, MAE, RMSE]

    
```

## Local model for Bioule (with past covariates)

Linear Regression Model with and without covariates

```python
from darts.models import LinearRegressionModel
```

```python
def fill_missing_values_of_series(series):
    for i in range(len(series)):
        series[i] = fill_missing_values(series[i])
    return series
```

```python
small_but_not_zero = 1e-3
df = compact_df_without_missing[compact_df_without_missing['model_name'] == 'saint_felix']
start_ts = pd.Timestamp('2020-06-01')
df = df[df["Date"].between("2000", "2020-12-01")]
df.loc[df['pluie'] == 0, 'pluie'] = small_but_not_zero
df.loc[df['etp'] == 0, 'etp'] = small_but_not_zero
series = TimeSeries.from_dataframe(df, 'Date', ['pz'], fill_missing_dates=True, freq='1d')
series = fill_missing_values(series)
```

```python
target = TimeSeries.from_dataframe(df, 'Date', ['pz'], fill_missing_dates=True, freq='1d')
cov = TimeSeries.from_dataframe(df, 'Date', ['pluie', 'etp', 'pompage'], fillna_value=0, fill_missing_dates=True, freq='1d')

[target, cov] = fill_missing_values_of_series([target, cov])

target_training, target_validation = target.split_before(start_ts)
cov_training, cov_validation = cov.split_before(start_ts)
target_training.plot()
target_validation.plot()
```

```python
?TimeSeries.from_dataframe
```

```python
cov_training.plot()
cov_validation.plot()
```

```python
#Linear regression without covariables
model_nocov = LinearRegressionModel(lags=30, output_chunk_length=180)
model_nocov.fit(series=target_training)

pred_nocov = model_nocov.predict(n=180, series=target_training)

target_training.plot(label='target')
target_validation.plot(label='truth')
pred_nocov.plot(label='pred')

print_metrics(target_validation, pred_nocov, target_training)
```

```python
#Linear regression with covariables

model_cov = LinearRegressionModel(lags=30, lags_past_covariates=30, output_chunk_length=180)
model_cov.fit(series=target_training, past_covariates=cov_training)

pred_cov = model_cov.predict(n=180, series=target_training, past_covariates=cov_training)

target_training.plot(label='target')
target_validation.plot(label='truth')
pred_cov.plot(label='pred')


print_metrics(target_validation, pred_cov, target_training)
```

### Linear regression 

```python
from darts.utils.missing_values import fill_missing_values
```

```python
dfs = [compact_df_without_missing[compact_df_without_missing['model_name'] == model] for model in models]
start_ts = pd.Timestamp('2020-06-01')

for df in dfs:
    df = df[df["Date"].between("2005", "2020-12-01")]
    
series = [TimeSeries.from_dataframe(df, 'Date', ['pz'], fill_missing_dates=True, freq='1d') for df in dfs]

series_train = []
series_valid = []
for serie in series:
    serie = fill_missing_values(serie)
    serie_training, serie_validation = serie.split_before(start_ts)
    series_train.append(serie_training)
    series_valid.append(serie_validation)
    
```

```python
#Linear regression without covariables global

model_nocov = LinearRegressionModel(lags=30, output_chunk_length=30)
model_nocov.fit(series=series_train)

```

```python
pred_nocov = model_nocov.predict(n=180, series=series_train)


metrics = np.zeros((6, 5))
for i in range(6):
    series_train[i].plot(label='target')
    series_valid[i].plot(label='truth')
    pred_nocov[i].plot(label='pred')
    metrics[i, :] = print_metrics(series_valid[i], pred_nocov[i], series_train[i])
    

    plt.show()
    

print(dict(zip(["MAPE", "sMAPE", "MASE", "MAE", "RMSE"], np.mean(metrics, axis=0))))
```
