```python
first_date = "2005-06-01"
end = pd.Timestamp('12-01-2022')

compact_df = pd.read_csv("compact_df.csv", index_col = 0)
compact_df['pz'] = compact_df['pz'].interpolate(method='polynomial', order=2)
compact_df_without_missing = compact_df[~compact_df['pz'].isnull()]
compact_df_without_missing['Date'] = pd.to_datetime(compact_df_without_missing['Date'], format='%Y-%m-%d')

# adding features
compact_df_without_missing['pluie_annuelle'] = compact_df_without_missing['pluie'].rolling(window=365).mean()
compact_df_without_missing['pluie_mensuelle'] = compact_df_without_missing['pluie'].rolling(window=30).mean()
compact_df_without_missing['pluie_hebdo'] = compact_df_without_missing['pluie'].rolling(window=30).mean()
compact_df_without_missing['etp_annuelle'] = compact_df_without_missing['etp'].rolling(window=365).mean()
compact_df_without_missing['etp_mensuelle'] = compact_df_without_missing['etp'].rolling(window=30).mean()
compact_df_without_missing['etp_hebdo'] = compact_df_without_missing['etp'].rolling(window=30).mean()

# remove zeros for covariates
small_but_not_zero = 1e-3
compact_df_without_missing.loc[compact_df_without_missing['pluie'] == 0, 'pluie'] = small_but_not_zero
compact_df_without_missing.loc[compact_df_without_missing['etp'] == 0, 'etp'] = small_but_not_zero

# subset of dates
compact_df_without_missing = compact_df_without_missing[
compact_df_without_missing["Date"].between(first_date, end)]
```

```python
models = compact_df.model_name.unique()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
for model in models:
    print(model)
    df = compact_df_without_missing[(compact_df_without_missing['model_name'] == model)]
    decomp = seasonal_decompose(df['pz'], model='additive', extrapolate_trend='freq', period=365)
    plt.plot(df['Date'], decomp.trend)
    plt.show()
    plt.plot(df['Date'], decomp.seasonal)
    plt.show()
```

```python
import numpy as np
years = ['2016', '2017', '2018', '2019', '2020', '2021']
scores = np.zeros((len(years), 5))
locations = ['bioule', 'les_barthes', 'saint_felix', 'saint_porquier', 'tarsac',
       'verniolles']
scores_global = np.zeros((len(locations), 5))

for id, location in enumerate(locations):
    for idx, year in enumerate(years):
        print(location, year)

        start_str = '06-01-' + year

        start = pd.Timestamp(start_str)
        end_str = '12-01-' + year

        features = ['pluie', 'etp', 'pompage', 'pluie_annuelle', 'pluie_mensuelle', 'etp_annuelle', 'etp_mensuelle']

        n_pred = len(pd.date_range(start_str, end_str))


        df = compact_df_without_missing[compact_df_without_missing['model_name'] == location]
        decomp = seasonal_decompose(df['pz'], model='additive', extrapolate_trend='freq', period=365)
        df['trend_pz'] = decomp.trend
        df['season_pz'] = decomp.seasonal
        season = TimeSeries.from_dataframe(df[(df['Date'].between(start_str, end_str)) & (df['model_name'] == location)], 'Date', ['season_pz'], fill_missing_dates=True, freq='1d')
        season = fill_missing_values(season)
        
        targets = TimeSeries.from_dataframe(df, 'Date', ['trend_pz', 'pz'], fill_missing_dates=True, freq='1d')
        covs = TimeSeries.from_dataframe(df, 'Date', features, fill_missing_dates=True, freq='1d')
        targets = fill_missing_values(targets)
        covs = fill_missing_values(covs)
        target_training, target_validation = targets.split_before(start)
        covs_training, covs_validation = covs.split_before(start)

        # Fit the model
        model_linear = LinearRegressionModel(lags=10, lags_past_covariates=20, output_chunk_length=n_pred)

        model_linear.fit(target_training['trend_pz'], covs_training)


        # Validation prediction
        pred_linear = model_linear.predict(n_pred)
        
        print(pred_linear.values().shape)
        print(season.values().shape)


        target_reconstruct = pred_linear.values() + season.values()
        print(target_reconstruct.shape)

        df_pred = pd.DataFrame({'Date': pd.date_range(start_str, end_str), 'pz_reconstructed': target_reconstruct[:, 0]})
        pred_reconstruct = TimeSeries.from_dataframe(df_pred, 'Date', 'pz_reconstructed', fill_missing_dates=True, freq='1d')

        # Visualize
        target_training['pz'].plot(label='Training')
        target_validation['pz'].plot(label='True')
        pred_reconstruct.plot(label='pred')


        from metrics import print_metrics
        score = print_metrics(target_validation['pz'], pred_reconstruct, target_training['pz'])
        scores[idx, :] = score
    scores_global[id, :] = np.mean(scores, axis=0)
    
print(dict(zip(["MAPE", "sMAPE", "MSE", "MAE", "RMSE"], np.mean(scores, axis=0))))


```
